```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


df = pd.read_csv('mtcars.csv')
```

Scatter plot

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x= df['mpg'], y=df['hp'], palette='viridis', data=df)
plt.title('Scatter Plot of MPG vs. Horsepower')
plt.xlabel('MPG')
plt.ylabel('Horsepower')
plt.show()
```

Histogram

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['mpg'], bins=10, kde=True)
plt.title('Histogram of MPG')
plt.show()
```

Bar plot

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=df['drat'], y=df['mpg'], data=df)
plt.title('Bar Plot')
plt.show()
```

Line chart

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=df['carb'], y=df['mpg'], data=df, marker='o')
plt.title('Line Chart')
plt.show()
```

Box plot

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['drat'], y=df['mpg'], data=df)
plt.title('Box Plot')
plt.show()
```

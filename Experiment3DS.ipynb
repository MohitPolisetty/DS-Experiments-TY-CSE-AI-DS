# Experiment 3

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder,OneHotEncoder
from sklearn.feature_extraction import FeatureHasher

df = pd.read_csv('student3.csv')
df1= pd.read_csv('Student_5.csv')
```

One hot Encoding

```python
one_hot_encoding = pd.get_dummies(df,columns=['marks '])
one_hot_encoding
```

Label Encoding

```python
label_encoder = LabelEncoder()
df1['encoded_name'] = label_encoder.fit_transform(df1['name'])
df1
```

Original Encoding

```python

sizes = pd.DataFrame({
    'Size': ['Small', 'Large', 'Medium', 'Large', 'Small']
})

categories = [['Small', 'Medium', 'Large']]

ordinal_encoder = OrdinalEncoder(categories=categories)
sizes['Size_encoded'] = ordinal_encoder.fit_transform(sizes[['Size']])
sizes
```

Count Encoding

```python
count_encoding = sizes['Size'].value_counts()
sizes['size_encode'] = sizes['Size'].map(count_encoding)
sizes
```

Hashing Encoding

```python
count_encoding = sizes['Size'].value_counts()
sizes['size_encode'] = sizes['Size'].map(count_encoding)
sizes
```
